## Correlation matrix to see if different numerical variables in the dataset correlate with each other


```python
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
data_numeric = data[numeric_cols]

corr_matrix = data_numeric.corr()

plt.figure(figsize=(8, 10))
sns.heatmap(corr_matrix, annot=True, cmap='YlOrRd')
plt.title('Correlation Matrix')
plt.show()
```

The correlation matrix plot shows the relationships between the numeric variables in the dataset.
The resulting correlation matrix provides insights into which variables are strongly correlated with each other. For example, it appears that **horsepower** and **engine size** are highly positively correlated.

## Which car model has the highest resale value compared to its initial price

```python
# Calculate the resale value percentage
data['resale_value'] = data['year_resale_value'] / data['price_in_thousands'] * 100

# Find the model with the highest resale value percentage
max_resale_model = data.loc[data['resale_value'].idxmax(), 'model']

print(f'The car model with the highest resale value compared to its initial price is: {max_resale_model}')
```
